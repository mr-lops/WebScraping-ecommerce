```python
# Importando bibliotecas
import requests    # Para fazer a conexao com a pagina
from bs4 import BeautifulSoup    # Para extrair elementos da pagina 
import re    # Para utilizar expressoes regulares 
import pandas as pd    # Para trabalhar com tabelas
```

```python
# Guardando o link da pagina em uma variavel
pagina = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
```

```python
# Solicitando a conexao com a pagina
resul = requests.get(pagina)
```

```python
# o resultado foi 200, quer dizer que a conexao foi bem sucedida
print(resul.status_code)
```

```python
# Pegando o conteudo da pagina e transformando em texto
codfonte = resul.text
codfonte
```

```python
# Filtrando somente o conteudo HTML da pagina
sopa = BeautifulSoup(codfonte, 'html.parser')
```

```python
# Vizualizando o conteudo da pagina mais organizado
print(sopa.prettify()) 
```

```python
#  Capturando todo HTML onde o conteudo dos tablets se encontram
blocos = sopa.find_all('div', class_='col-sm-4 col-lg-4 col-md-4')
type(blocos)
```

```python
# Função para filtrar as informações dos tablets
def filtraDados(dat):
    
    # nome do produto
    nome = dat.find('a', class_='title').text
    
    # preço do produto
    precoStr = dat.find('h4', class_='pull-right price').text
    preco = float(precoStr.replace('$',''))
    
    # Descrição do produto
    des = dat.find('p', class_='description').text
    
    # Classificação do produto
    clasf = len(dat.find_all('span',  class_='glyphicon glyphicon-star'))
    quantOp = int((re.findall('\d', (dat.find('p', class_='pull-right').text)))[0])
    
    # transformando o conteudo em formato de dicionario
    conteudo = {'Nome': nome, 'Descricao': des, 'Classificacao': clasf, 'Quant_Opinioes': quantOp, 'Valor': preco}
    
    # retornando as informações do tablet
    return conteudo
```

```python
# criando uma lista para armazenar as informações dos tablets
produtos = []
```

```python
# FOR para percorrer cada bloco HTML de um tablet para filtrar o conteudo e armazenar na variavel produtos
for blc in blocos:
    dadoTratado = filtraDados(blc)
    produtos.append(dadoTratado)
```

```python
# cada informação do tablet esta em formato dicionario e todos estao na lista produtos
produtos[0]
```

```python
# Utilizando o Pandas para transformar a lista com as informações dos tablets em uma tabela
tablets = pd.DataFrame(produtos)
tablets
```

```python
# exportando a tabela para o formato CSV
tablets.to_csv('Tablets.csv', encoding = "utf-8", index = False)
```
